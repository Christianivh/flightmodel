```python
import pandas as pd
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
from matplotlib import cm

import itertools

flightdata = pd.read_csv('datos.csv',sep=';')
flightdata.head(5)
```

```python
flightdata.shape
```

```python
flightdata = flightdata.fillna(flightdata.median())
flightdata.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import pickle



unit = set(flightdata['ID'])
maxtam = len(unit)
maxtam

s = []

for i in range(1,maxtam):
    print("Ejecutando ",list(unit)[i])
    mvalor = list(unit)[i]
    Mod = flightdata[(flightdata['ID'] == mvalor)]
    entrenamiento = Mod[(Mod['filtro'] == 1)]
    testing = Mod[(Mod['filtro'] == 0)]
    testing.sort_values(by=['aniomes'])
    
    ## creacion de entrenamiento
    
    X_train = entrenamiento[(['arr_flights', 'carrier_ct', ' weather_ct','nas_ct'])]
    y_train = entrenamiento[(['target_vuelos_de_retraso'])]
    
    ## creacion de testing
    X_test= testing[(['arr_flights', 'carrier_ct', ' weather_ct','nas_ct'])]
    y_test = testing[(['target_vuelos_de_retraso'])]
    
    #### modelo 
    regr = LinearRegression()
    regr.fit(X_train.values, y_train.values)
    y_pred = regr.predict(X_test.values)
    pickle.dump( regr, open('d:/Users/gnunuvero/Desktop/modelos_vuelos/' + mvalor + '.pkl', 'wb'))
    #ECM = mean_squared_error(y_test.values, y_pred)
    s.append(y_pred)
    # The coefficients
    

 
```
